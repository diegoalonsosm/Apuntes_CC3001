*Existe un algoritmo alternativo a Hoare, que resulta en una codificación más 
sencilla. Este algoritmo, debido a Lomuto, se basa en el siguiente invariante:*

*En este algoritmo, en cada iteración, si  a[j]<p , se intercambian  a[i]  con a[j]  y se incrementa  i , porque ahora hay un elemento más en el grupo de los menores que  p . Después de esto, se incrementa  j , incondicionalmente 
(¿por qué es correcto hacer eso?).*

*Programe la partición de Lomuto en el recuadro siguiente y pruébela.* \\

```python
# particionLomuto: list int -> tuple
# Entrega una tupla ( p = punto de corte, i = elementos <p, a = lista particionada )
# Ej: assert particionLomuto(A,100) -> ([73,21,34,98,56,37,77,65,82,15,36],100,[]) 
# Ej: particionLomuto(A,50) -> (50,5,[21,34,37,15,36,73,77,65,82,98,56])

def particionLomuto(a,p):
    
    # Se ejecuta hasta que no hayan elementos desconocidos, es decir, 
    # hasta que se haya identificado la relación de orden que tiene p
    # con cada uno de los elemento de la lista
    
    # Asignación de variables
    n=len(a); i=0; j=0
    
    while j<=n-1: # Condición: quedan elementos desconocidos
    
        if a[j]<p:
            (a[i],a[j])=(a[j],a[i]) # Se intercambian los elementos en i y j
            i+=1
        j+=1

    return (p,i,a)

# Lista para ser particionada
A = [73,21,34,98,56,37,77,65,82,15,36]

particionLomuto(A,50)

# Pruebas
assert particionLomuto(A,50) == (50,5,[21,34,37,15,36,73,77,65,82,98,56])
assert particionLomuto(A,100) == (100,11,[21,34,37,15,36,73,77,65,82,98,56])
```
