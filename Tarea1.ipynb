**Regla:** *Se escriben separados por yuxtapusición (sin el conector "y").* \\
**Ejemplo:** *ciento dos, mil cuatrocientos treinta, trescientos mil veintiuno.* \\

### **Código para la función** ***palabras(n)***
La idea principal para la escritura de este código se basa en escribir (de izquierda a derecha) los cardinales, aplicando las reglas citadas anteriormente. Para lograr esto, se utiliza un bloque 'for' (para escribir el número final) y una función auxiliar (que se encarga de las centenas, decenas y unidades de cada ciclo de escritura, *i.e. cdu de millón, cdu de mil, cdu normales*). Cada iteración del bloque 'for' se referencia con los *ciclos i*. Dicha nomenclatura se explica a lo largo del código.

```python
# palabras: int -> str
# Entrega una cadena de texto con el cardinal del número ingresado.
# Los ejemplos de esta función se encuentran en el bloque de texto siguiente 

def palabras(n):
    assert 0<=n<=999999999

    # Función auxiliar: genera las centenas, decenas y unidades de cada ciclo
    def _palabras(n,i):
        
        """""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""
        Importante:
        La variable i se utiliza para indicar el ciclo de escritura.

        - i = 0 -> cdu de millón
        - i = 1 -> cdu de mil
        - i = 2 -> cdu

        *cdu: abreviación para centenas decenas y unidades

        """""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""""
        # Se salta a las siguientes tres cifras
        if n == 0: return [''] # También se podría usar None
    
        # Excepción: 'cien' y no 'ciento' como aparece en lista decenas
        if n == 100: return ['cien']

        # Tratamiento de los cardinales simples 
        unidades = ['','uno' if i==2 else 'un','dos','tres','cuatro', 
                    'cinco','seis','siete','ocho','nueve']
    
        decenas = ['','diez','veinte','treinta','cuarenta',
                   'cincuenta','sesenta','setenta','ochenta','noventa']
    
        centenas = ['','ciento','doscientos','trescientos','cuatrocientos',
                    'quinientos','seiscientos','setecientos','ochocientos','novecientos']
    
        # Tratamiento de los cardinales compuestos simples y excepciones varias (11-19,21-29)
        excepciones = {11:'once',12:'doce',13:'trece',14:'catorce',15:'quince',
                       16:'dieciséis',17:'diecisiete',18:'dieciocho',19:'diecinueve',
                       21:'veintiuno' if i==2 else 'veintiún',22:'veintidós',23:'veintitrés',
                       24:'veinticuatro',25:'veinticinco',26:'veintiséis',27:'veintisiete',
                       28:'veintiocho',29:'veintinueve'}
    
        (c,d,u) = (n//100,(n%100//10),n%10); du = int(str(d)+str(u)) # número entre 0-99
        
        # Excepción: evitar el escribir 'un mil' (i=1, ciclo cdu de mil)
        if n==1 and i==1:
            return ['']
        
        # Excepción: múltiplos de diez
        elif du%10==0:
            return [centenas[c],decenas[d]]
    
        # Excepción: cardinales compuestos simples [16-19,21-29]
        elif 11<=du<=19 or 21<=du<=29:
            return [centenas[c],excepciones[du]]
        
        # Excepción: múltiplos de diez
        elif d==0:
            return [centenas[c],unidades[u]]

        # Manera estándar de escribir un cardinal
        else:
            return [centenas[c],decenas[d],'y',unidades[u]]
    
    # Comienzo de función principal

    # Caso 'trivial'
    if n==0:
        return 'Cero'
    
    # Separación cdu de millones, cdu de miles y cdu
    (a,b,c)=(n//1000000,(n%1000000)//1000,n%1000)

    palabra = [''] # Base para la construcción del cardinal
    
    for i in range(0,3):
        
        # i = 0 -> cdu de millón
        # i = 1 -> cdu de mil
        # i = 2 -> cdu
        
        cifras = a if i==0 else (b if i==1 else c)
        
        if cifras!=0 and not (a==1 and i==0):
            separador = 'millones' if i==0 else ('mil' if i==1 else '')
            palabra += _palabras(cifras,i) + [separador]
        
        # Excepción 'un millón' y no 'un millones'
        elif cifras!=0:
            palabra += ['un millón']
    
    # Elimina los '' residuales en la lista para usar el método join() correctamente
    while '' in palabra: palabra.remove('')
    
    return (' '.join(palabra)).capitalize()
```

### **Ejemplos para la función** ***palabras(n)***

```python
# Formateo para ejemplos (ignorar)

def test(n):
    (a,b,c)=(n//1000000,(n%1000000)//1000,n%1000)

    a = "{:03d}".format(a)
    b = "{:03d}".format(b)
    c = "{:03d}".format(c)

    print(f'- ({a} {b} {c}) {palabras(n)}')

print('===================')
print(' Lista de ejemplos ')
print('===================\n')

print('1) Se "saltan" ciclos')
test(111) # 000000111
test(111000) # 000111000
test(111000000) # 111000000
test(111111) # 000111111
test(111000111) # 111000111
test(111111000) # 111111000
test(111111111) # 111111111
print()

print('2) Cero, mil y un millón')
test(0)
test(1000)
test(1000000)
print()

print('3) Cifras terminadas con "un"')
test(21021021)
test(31031031)
print()

print('4) Múltiplos de diez')
test(110)
test(110110)
test(110110110)
print()

print('5) Cardinales compuestos simples y coordinados')
test(19019019)
test(25025025)
test(41041041)
print()

test(10000000)
```

```python
from random import randint

# Aplica la función palabras() para números aleatorios en los intervalos SIGNIFICATIVOS.

def test1():
    print('Lista de prueba:')
    intervalos=[9,99,999,9999,99999,999999,9999999,99999999,999999999]
    for vuelta in intervalos:
        alAzar = randint(0,vuelta)
        (a,b,c)=(alAzar//1000000,(alAzar%1000000)//1000,alAzar%1000)
        numero = palabras(alAzar)
        
        a = "{:03d}".format(a)
        b = "{:03d}".format(b)
        c = "{:03d}".format(c)
        
        print(f'- ({a} {b} {c}) {numero}')
        
test1()
```
